## Normalize data
1. Find the max and min values for each column
1. divide each row's attribute minus the min attribute by the max attribute minus the min attribute
1. This will set all attributes to values between 0 and 1
1. better for neural network

```python
for i in df:
    df[i] = (df[i] - df[i].max())/(df[i].min()- df[i].max())
```
